## Make MA_{ma} columns

```python
intervals = ['1h', '2h', '4h', '12h', '1d']
for interval in intervals:
    btc_data = pd.read_csv(f'data/btc_all_{interval}_candles_from_2017.csv')
    ma_candidates = [10, 20, 30, 50, 60, 90, 100, 110, 120, 130, 140, 150, 160, 170, 180, 200]
    # Calculate moving averages for each candidate
    for ma in ma_candidates:
        btc_data[f'MA_{ma}'] = btc_data['close'].rolling(window=ma).mean()
    btc_data.to_csv(f'data/btc_{interval}_data_with_ma.csv', index=False)
```
