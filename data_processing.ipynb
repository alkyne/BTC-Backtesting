```python
import pandas as pd
import numpy as np
```

```python
btc_data = pd.read_csv('btc_all_1h_candles_from_2017.csv')
```

```python
btc_data.head()
```

## Make MA_{ma} columns

```python
ma_candidates = [10, 20, 30, 50, 60, 90, 100, 110, 120, 130, 140, 150, 160, 170, 180, 200]
# Calculate moving averages for each candidate
for ma in ma_candidates:
    btc_data[f'MA_{ma}'] = btc_data['close'].rolling(window=ma).mean()
```

```python

# Display the first few rows of the updated DataFrame
print(btc_data.head(50))
```

```python
btc_data.columns
```

```python
btc_data.to_csv('btc_1h_data_with_ma.csv', index=False)
```

## position && signal

```python
df = pd.read_csv('btc_1h_data_with_ma.csv')
```

```python
df.head(10)
```

```python
ma = 20
```

```python
df['Position'] = np.where(df['close'] > df[f'MA_{ma}'], 1, 0)  # 1 for buy, 0 for sell
df['Signal'] = df['Position'].diff()  # Buy signal when diff > 0, sell signal when diff < 0
```

```python
df[['timestamp', 'close', f'MA_{ma}', 'Position', 'Signal']].to_csv('tmp_btc_1h_ma_20_data_with_position_signal.csv', index=False)
```
